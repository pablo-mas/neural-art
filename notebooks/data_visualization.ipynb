```python
import seaborn as sns
sns.set_theme(style="darkgrid")
import matplotlib.pyplot as plt

from neuralart.data import *
```

# Data Visualization

## Load the data

```python
data = pd.read_csv("../raw_data/wikiart/new_dataset/wikiart-target_style-class_27.csv")
```

## Number of classes per target

```python
col = ['artist','genre','style']

fig, ax = plt.subplots(1,1,figsize=(10,5))
sns.barplot(y=data[col].nunique().index, 
            x=data[col].nunique().values,
            order=data[col].nunique().sort_values(ascending=False).index,
            ax=ax);

for i, v in enumerate(data[col].nunique().sort_values(ascending=False).values):
    ax.text(v, i , str(v), color='blue', fontweight='bold')
    
ax.set_title("Number of classes per target");
ax.set_xlabel("Count");
ax.set_ylabel("Target");
```

## Number of images per style

```python
fig, ax = plt.subplots(1,1,figsize=(15,15));
sns.countplot(data=data, y="style", 
              order = data['style'].value_counts().index,
              ax=ax);
sns.despine()

for i, v in enumerate(data["style"].value_counts().values):
    ax.text(v, i, str(v), color='blue')
    
ax.set_title("Number of images per style");
```

## Number of images per genre

```python
fig, ax = plt.subplots(1,1,figsize=(15,10));
sns.countplot(data=data, y="genre", 
              order = data['genre'].value_counts().index,
              ax=ax);
sns.despine()

for i, v in enumerate(data["genre"].value_counts().values):
    ax.text(v, i , str(v), color='blue', fontweight='bold')
    
ax.set_title("Number of images per genre");
```

## Number of artists per movement

```python
fig, ax = plt.subplots(1,1,figsize=(15,10));
sns.barplot(y=data.groupby("style").artist.nunique().index, 
            x=data.groupby("style").artist.nunique().values,
            order=data.groupby("style").artist.nunique().sort_values(ascending=False).index,
            ax=ax);

for i, v in enumerate(data.groupby("style").artist.nunique().sort_values(ascending=False).values):
    ax.text(v, i , str(v), color='blue', fontweight='bold')
    
ax.set_title("Number of artists per style");
```
