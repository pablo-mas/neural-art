# Import

```python
%load_ext autoreload
%autoreload 2
```

```python
from neuralart.trainer import Trainer
```

# Initialize the trainer

```python
trainer = Trainer()
```

# Create TensorFlow Dataset

```python
batch_size = 32
img_height = 224
img_width = 224

image_folder_path = "../raw_data/wikiart/wikiart-target_style-class_8-keepgenre_True-merge_style_m1-n_100_max-flat_False"
csv_path = "../raw_data/wikiart/wikiart-target_style-class_8-keepgenre_True-merge_style_m1-n_100_max-flat_False.csv"
```

## Create TensorFlow Dataset using tf.keras.image_dataset_from_directory() method

```python
trainer.create_dataset_from_directory(image_folder_path, batch_size, img_height, img_width)
```

```python
trainer.plot_train_batch()
```

## Create TensorFlow Dataset using tf.data.Dataset.from_tensor_slices() method

```python
trainer.create_dataset_from_csv(csv_path, image_folder_path, batch_size, img_height, img_width, shuffle_dataframe=True)
```

```python
trainer.plot_train_batch()
```

# Build VGG16 Model

```python
trainer.build_model("VGG16", trainable_layers=2, random_roration=0.3, random_zoom=0.3, learning_rate=0.001)
```

# Train Model

```python
trainer.run(epochs=5)
```

```python
trainer.plot_history()
```

# Load an existing model

```python
model_path = "../models/VGG16/20210831-135428-images_41472-unfreeze_2-batch_128/"
trainer.load_model(model_path)
```

# Evaluate the Model

```python
trainer.plot_confusion_matrix()
```

```python
trainer.evaluate()
```
